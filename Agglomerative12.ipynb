```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Generate a synthetic dataset
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)
```

```python
# Step 1: Perform Agglomerative Clustering
# Perform Agglomerative Clustering
agg_clust = AgglomerativeClustering(n_clusters=3)
y_pred = agg_clust.fit_predict(X)
```

```python
# Plotting the dendrogram
plt.figure(figsize=(15, 10))
linked = linkage(X, 'ward')
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,
            truncate_mode='lastp',
           p=20)
           # show only the last 12 merged clusters

plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
plt.figure(figsize=(15, 10), dpi=150)  # Increased figure size and DPI
linked = linkage(X, 'ward')
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True,
           leaf_rotation=90,  # Rotated x-axis labels
           leaf_font_size=8, # Reduced font size for x-axis labels
           truncate_mode='lastp',  # show only the last p merged clusters
           p=12  # show only the last 12 merged clusters
          )
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
# Plotting the clusters
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='rainbow')
plt.title('Agglomerative Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
